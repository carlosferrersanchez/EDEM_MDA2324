```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs # Importar clase para crear clusters ficticios con sklearn

# Tarea: crear 3 clusters, con 2 características, la desviación del cluster que sea 0.5
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.5, random_state=0) # añadir aquí la definición del dataset

# Plotear los puntos con matplotlib
plt.scatter(X[:, 0], X[:, 1])
plt.title("Puntos iniciales")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

```python
# Entrenamos el algoritmo kmeans al dataset y obtenemos la predicción.
# Luego entenderemos como funciona

# 1. Importamos el modelo
from sklearn.cluster import KMeans

# 2. Entrenamos el modelo
model = KMeans(n_clusters=3, random_state=0)
model.fit(X)

# 3. Predecimos
y_kmeans = model.predict(X)
```

```python
# Tarea: Obtener el "accuracy" del modelo.
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true, y_kmeans)
print(accuracy)
```

En este caso, podemos sacar el accuracy. ¿Podremos en un dataset real de clustering?

```python
# Tarea: Obtenemos los centros de los clusters
centers = model.cluster_centers_

print("Centros de los clusters:")
print(centers)
```

# ¿Algo interesante respecto a los centros?

```python
# Plotear los mismos puntos que anteriormente, pero con el color de los puntos según su cluster,
# mirad "c"
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Tarea: plotear los centros
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

plt.suptitle("3 grupos aleatorios de puntos - los centros de los clusters fueron encontrados usando el algoritmo K-means")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```
