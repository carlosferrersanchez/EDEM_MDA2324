Como los valores estan en una escala 0,1 solamente tendremos que codificar las etiquetas.

```
from sklearn.preprocessing import OneHotEncoder
```

```
# Convert pandas Dataframe to Numpy Array
data_numpy = data.values
```

```
# Show data type before and after conversion
print(type(data))
print(type(data_numpy))
```

```
data_numpy.shape
```

```
# Separate input variables from output label
X = data_numpy[:, :60]
y = data_numpy[:, 60:61]
```

```
# Show shape of training data and labels
print("Shape data: ", X.shape)
print("Shape labels: ", y.shape)
```

```
# One-hot encoding of labels
from sklearn.preprocessing import OneHotEncoder
onehot_enc = OneHotEncoder()
y_onehot = onehot_enc.fit_transform(y).toarray()
```

```
# Obtiene las categorías (codificaciones) de las etiquetas
codificaciones_etiquetas = onehot_enc.categories_

# Imprime las codificaciones de las etiquetas
print("Codificaciones de las etiquetas:")
print(codificaciones_etiquetas)
```

```
y_onehot = y_onehot.astype('float32')
X = X.astype('float32')
```

```
# Split data in training and validation partitions
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y_onehot,
                                                  test_size=0.1)
```

```
# Import dependencies for designing Keras model
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```
# Design simple neural network architecture
model = Sequential()
model.add(Dense(60, input_dim=60, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```
# Show model information
model.summary()
```

```
# Choose loss function, optimizer and training metrics
model.compile(loss='binary_crossentropy', optimizer='adam',
              metrics=['accuracy'])
```

```
# Choose number of epochs and batch size and train the model
history = model.fit(X_train, y_train, epochs=84, batch_size=60,
                    validation_data=(X_val, y_val))
```
