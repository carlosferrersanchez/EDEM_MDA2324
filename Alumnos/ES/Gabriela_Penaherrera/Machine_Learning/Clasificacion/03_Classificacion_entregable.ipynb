```python
from sklearn.model_selection import train_test_split

## A COMPLETAR

'''Crear los sets de entrenamiento (70%), validación (15%)
    y prueba (15%)'''

def crear_train_validation_test (x,y):

  xtr, x_rest, ytr, y_rest = train_test_split(x, y, test_size=0.3, random_state=43)
  xvl, xts, yvl, yts = train_test_split(x_rest, y_rest, test_size=0.5, random_state=43)

  return xtr, ytr, xvl, yvl, xts, yts


# Ejecutar la función

x_train, y_train, x_val, y_val, x_test, y_test = crear_train_validation_test(x,y)

# Verificar
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
print(x_test.shape, y_test.shape)
```

```python
def escalar_datos(xtr, xvl, xts):
    '''Escalar datos de entrada (entrenamiento, validación y prueba)
    en el rango de 0 a 1'''
    scaler = MinMaxScaler()

    # Fit + transform sobre el set de entrenamiento
    xtr_s = scaler.fit_transform(xtr)

    # Transform sobre prueba y validación
    xvl_s = scaler.transform(xvl)
    xts_s = scaler.transform(xts)

    return xtr_s, xvl_s, xts_s

# Ejecutar la función
x_train_s, x_val_s, x_test_s = escalar_datos(x_train, x_val, x_test)

# Verificar
print(x_train_s.min(axis=0),x_train_s.max(axis=0))
print(x_val_s.min(axis=0),x_val_s.max(axis=0))
print(x_test_s.min(axis=0),x_test_s.max(axis=0))
```

#Modelos

Árbol de clasificación:

```python
## A COMPLETAR

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

def entrenar_y_evaluar_arbol(x_train, y_train, x_val, y_val, max_depth=None):
    '''Entrena y evalúa un árbol de clasificación con los datos proporcionados'''
    # Inicializar el clasificador de árbol de decisión
    clf = DecisionTreeClassifier(max_depth=max_depth, random_state=42)

    # Entrenar el modelo con los datos de entrenamiento
    clf.fit(x_train, y_train)

    # Predecir las etiquetas de clase en los datos de validación
    y_pred_val = clf.predict(x_val)

    # Calcular la precisión en los datos de validación
    accuracy = accuracy_score(y_val, y_pred_val)
    puntaje_f1 = f1_score(y_val, y_pred_val, average='weighted')  # 'macro' es útil cuando tienes clases desbalanceadas

    return clf, accuracy, puntaje_f1


# Entrenar y evaluar un árbol de clasificación
arbol, precision, puntaje_f1 = entrenar_y_evaluar_arbol(x_train_s, y_train, x_val_s, y_val)

# Imprimir la precisión obtenida
print("Precisión del árbol de clasificación:", precision)
print("Precisión del árbol de clasificación f1_score:", puntaje_f1)

```

Clasificador kNN:

```python
## A COMPLETAR

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def entrenar_y_evaluar_knn(x_train, y_train, x_val, y_val, n_neighbors=5):
    '''Entrena y evalúa un clasificador kNN con los datos proporcionados'''
    # Inicializar el clasificador kNN
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)

    # Entrenar el modelo con los datos de entrenamiento
    clf.fit(x_train, y_train)

    # Predecir las etiquetas de clase en los datos de validación
    y_pred_val = clf.predict(x_val)

    # Calcular la precisión en los datos de validación
    accuracy = accuracy_score(y_val, y_pred_val)
    puntaje_f1 = f1_score(y_val, y_pred_val, average='weighted')

    return clf, accuracy, puntaje_f1

# Entrenar y evaluar un clasificador kNN
knn, precision_knn, puntaje_f1 = entrenar_y_evaluar_knn(x_train_s, y_train, x_val_s, y_val)

# Imprimir la precisión obtenida
print("Precisión del clasificador kNN:", precision_knn)
print("Precisión kNN f1_score:", puntaje_f1)
```

```python
from sklearn.svm import SVC

def entrenar_y_evaluar_svm(x_train, y_train, x_val, y_val, kernel='rbf', C=1.0):
    '''Entrena y evalúa un modelo de SVM con los datos proporcionados'''
    # Inicializar el clasificador SVM
    clf = SVC(kernel=kernel, C=C, random_state=42)

    # Entrenar el modelo con los datos de entrenamiento
    clf.fit(x_train, y_train)

    # Predecir las etiquetas de clase en los datos de validación
    y_pred_val = clf.predict(x_val)

    # Calcular la precisión en los datos de validación
    accuracy = accuracy_score(y_val, y_pred_val)
    puntaje_f1 = f1_score(y_val, y_pred_val, average='weighted')

    return clf, accuracy, puntaje_f1


# Entrenar y evaluar un modelo SVM con escalamiento
svm_con_escalamiento, precision_svm_con_escalamiento, puntaje_f1 = entrenar_y_evaluar_svm(x_train_s, y_train, x_val_s, y_val)

# Imprimir la precisión obtenida
print("Precisión del modelo SVM con escalamiento:", precision_svm_con_escalamiento)
print("Precisión del modelo SVM con f1score:", puntaje_f1)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

def entrenar_y_evaluar_random_forest(x_train, y_train, x_val, y_val, n_estimators=100, max_depth=None, random_state=42):
    '''Entrena y evalúa un modelo de Random Forest con los datos proporcionados'''
    # Inicializar el clasificador Random Forest
    clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=43)

    # Entrenar el modelo con los datos de entrenamiento
    clf.fit(x_train, y_train)

    # Predecir las etiquetas de clase en los datos de validación
    y_pred_val = clf.predict(x_val)

    # Calcular la precisión en los datos de validación
    accuracy = accuracy_score(y_val, y_pred_val)
    puntaje_f1 = f1_score(y_val, y_pred_val, average='weighted')

    return clf, accuracy, puntaje_f1

# Entrenar y evaluar un modelo Random Forest con escalamiento
random_forest_con_escalamiento, precision_random_forest_con_escalamiento, puntaje_f1 = entrenar_y_evaluar_random_forest(x_train_s, y_train, x_val_s, y_val)

# Imprimir la precisión obtenida
print("Precisión del modelo Random Forest con escalamiento:", precision_random_forest_con_escalamiento)
print("Precisión del modelo Random Forest con f1score:", puntaje_f1 )
```

#ENSEMBLES

```python
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score

def train_and_evaluate_classifier(clf, x_train, y_train, x_val, y_val):
    '''Entrena y evalúa un clasificador con los datos proporcionados'''
    # Entrenar el clasificador
    clf.fit(x_train, y_train)

    # Predecir las etiquetas de clase en los datos de validación
    y_pred_val = clf.predict(x_val)

    # Calcular la precisión en los datos de validación
    accuracy = accuracy_score(y_val, y_pred_val)

    # Calcular el puntaje F1 en los datos de validación
    f1 = f1_score(y_val, y_pred_val, average='weighted')

    return clf, accuracy, f1

# Crear los clasificadores individuales
knn_classifier = KNeighborsClassifier()
svm_classifier = SVC(probability=True)  # Habilita las probabilidades para VotingClassifier
nb_classifier = GaussianNB()
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42)

# Crear el VotingClassifier con los modelos específicos
ensemble = VotingClassifier(estimators=[
    ('knn', knn_classifier),
    ('svm', svm_classifier),
    ('nb', nb_classifier),
    ('rf', rf_classifier)
], voting='hard')  # Cambia a 'soft' si deseas utilizar votación suave

# Entrenar y evaluar el ensemble
ensemble, precision_ensemble, f1_ensemble = train_and_evaluate_classifier(ensemble, x_train_s, y_train, x_val_s, y_val)

# Imprimir la precisión y el puntaje F1 obtenidos
print("Precisión del modelo Voting Classifier:", precision_ensemble)
print("Puntaje F1 del modelo Voting Classifier:", f1_ensemble)
```
